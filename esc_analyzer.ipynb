```python
import urllib.request
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


class ESCAnalyzer:
    def __init__(self, von_jahr, bis_jahr):
        self.df_live = self._fetch_api_data(von_jahr, bis_jahr)
        self.ergebnis = self._berechne_bias()
        self.land_mapping = {
            'AL': ('albanien', 'albania'), 'AM': ('armenien', 'armenia'), 'AT': ('österreich', 'austria'),
            'AU': ('australien', 'australia'), 'AZ': ('aserbaidschan', 'azerbaijan'), 'BA': ('bosnien', 'bosnia'),
            'BE': ('belgien', 'belgium'), 'BG': ('bulgarien', 'bulgaria'), 'BY': ('weißrussland', 'belarus'),
            'CH': ('schweiz', 'switzerland'), 'CY': ('zypern', 'cyprus'), 'CZ': ('tschechien', 'czechia'),
            'DE': ('deutschland', 'germany'), 'DK': ('dänemark', 'denmark'), 'EE': ('estland', 'estonia'),
            'ES': ('spanien', 'spain'), 'FI': ('finnland', 'finland'), 'FR': ('frankreich', 'france'),
            'GB': ('großbritannien', 'united kingdom'), 'GE': ('georgien', 'georgia'), 'GR': ('griechenland', 'greece'),
            'HR': ('kroatien', 'croatia'), 'HU': ('ungarn', 'hungary'), 'IE': ('irland', 'ireland'),
            'IL': ('israel', 'israel'), 'IS': ('island', 'iceland'), 'IT': ('italien', 'italy'),
            'LT': ('litauen', 'lithuania'), 'LU': ('luxemburg', 'luxembourg'), 'LV': ('lettland', 'latvia'),
            'MD': ('moldawien', 'moldova'), 'ME': ('montenegro', 'montenegro'), 'MK': ('nordmazedonien', 'north macedonia'),
            'MT': ('malta', 'malta'), 'NL': ('niederlande', 'netherlands'), 'NO': ('norwegen', 'norway'),
            'PL': ('polen', 'poland'), 'PT': ('portugal', 'portugal'), 'RO': ('rumänien', 'romania'),
            'RS': ('serbien', 'serbia'), 'RU': ('russland', 'russia'), 'SE': ('schweden', 'sweden'),
            'SI': ('slowenien', 'slovenia'), 'SM': ('san marino', 'san marino'), 'UA': ('ukraine', 'ukraine'),
            'WLD': ('welt', 'rest of the world')
        }
    def _fetch_api_data(self, von_jahr, bis_jahr):
        alle_votes = []
        for jahr in range(von_jahr, bis_jahr + 1):
            if jahr == 2020:
                continue
            url = f"https://eurovisionapi.runasp.net/api/senior/contests/{jahr}"
            try:
                req = urllib.request.Request(
                    url,
                    headers={'User-Agent': 'Mozilla/5.0'}
                )
                with urllib.request.urlopen(req) as response:
                    data = json.loads(response.read().decode())
                    id_zu_land = {
                        c['id']: c['country']
                        for c in data.get('contestants', [])
                        if 'id' in c and 'country' in c
                    }
                    for runde in data.get('rounds', []):
                        if 'final' in runde.get('name', '').lower():
                            for perf in runde.get('performances', []):
                                receiver_land = id_zu_land.get(
                                    perf.get('contestantId')
                                )
                                if not receiver_land:
                                    continue
                                for score in perf.get('scores', []):
                                    if score.get('name') == 'public':
                                        for voter_land, punkte in score.get(
                                            'votes', {}
                                        ).items():
                                            alle_votes.append({
                                                'Voter': str(voter_land).strip(),
                                                'Receiver': str(receiver_land).strip(),
                                                'Punkte': int(punkte)
                                            })
            except Exception as e:
                print(f"Fehler bei Jahr {jahr}: {e}")
                continue
        return pd.DataFrame(alle_votes)
        
    def _berechne_bias(self):
        if self.df_live.empty:
            return pd.DataFrame()
        punkte = (
            self.df_live
            .groupby(['Voter', 'Receiver'])['Punkte']
            .sum()
            .reset_index()
        )
        jahre = (
            self.df_live
            .groupby(['Voter', 'Receiver'])['Punkte']
            .count()
            .reset_index()
            .rename(columns={'Punkte': 'Jahre_Aktiv'})
        )
        df = pd.merge(
            punkte,
            jahre,
            on=['Voter', 'Receiver']
        )
        df['Prozent_Erhalten'] = ( df['Punkte'] / ( df['Jahre_Aktiv'] * 12 ) ) * 100
        schnitt = ( ( df.groupby('Receiver')['Punkte'].sum() / df.groupby('Receiver')['Jahre_Aktiv'].sum() / 12 * 100 ).reset_index().rename(columns={0: 'Globaler_Schnitt'}) )
        df = pd.merge(df, schnitt, on='Receiver')
        df['Nachbarschafts_Bias'] = ( (df['Prozent_Erhalten'] + 0.01) / (df['Globaler_Schnitt'] + 0.01) )
        return ( df[df['Jahre_Aktiv'] >= 3].sort_values( by='Nachbarschafts_Bias', ascending=False ).reset_index(drop=True) )
        
    def _normalisiere_land(self, eingabe):
        s = str(eingabe).strip().lower()
        if s.upper() in self.land_mapping:
            return s.upper()
        for iso, namen in self.land_mapping.items():
            if s == namen[0] or s == namen[1]:
                return iso
        raise ValueError(
            f"Land '{eingabe}' wurde nicht gefunden!"
        )
        
    def zeige_paar_bias(self, voter, receiver):
        try:
            v_iso = self._normalisiere_land(voter)
            r_iso = self._normalisiere_land(receiver)
        except ValueError as e:
            print(e)
            return
        zeile = self.ergebnis[ (self.ergebnis['Voter'] == v_iso) & (self.ergebnis['Receiver'] == r_iso) ]
        if zeile.empty:
            print( f"Keine ausreichenden Daten " f"für {v_iso} -> {r_iso}" )
            return
        rang = zeile.index[0] + 1
        bias = zeile['Nachbarschafts_Bias'].values[0]
        gesamt = len(self.ergebnis)
        print(f"Paarung: {v_iso} ➔ {r_iso}")
        print(f"Rang: {rang} von {gesamt}")
        print(f"Bias: {bias:.2f}x")

    def zeige_einseitigkeit(self, land_a, land_b):
        try:
            a = self._normalisiere_land(land_a)
            b = self._normalisiere_land(land_b)
        except ValueError as e:
            print(e)
            return
        ab = self.ergebnis[ (self.ergebnis['Voter'] == a) & (self.ergebnis['Receiver'] == b) ]
        ba = self.ergebnis[ (self.ergebnis['Voter'] == b) & (self.ergebnis['Receiver'] == a) ]
        if ab.empty or ba.empty:
            print("Nicht genügend Daten vorhanden.")
            return
        bias_ab = ab['Nachbarschafts_Bias'].values[0]
        bias_ba = ba['Nachbarschafts_Bias'].values[0]
        asymmetrie = ( max(bias_ab, bias_ba) / min(bias_ab, bias_ba) )
        if bias_ab > bias_ba:
            dominante_richtung = f"{a} ➔ {b}"
        else:
            dominante_richtung = f"{b} ➔ {a}"
        print(f"Voting-Beziehung: {a} ↔ {b}")
        print(f"{a} ➔ {b}: {bias_ab:.2f}x")
        print(f"{b} ➔ {a}: {bias_ba:.2f}x")
        print("\nEinseitigkeits-Index:")
        print(f"{asymmetrie:.2f}x")
        if asymmetrie < 1.3:
            interpretation = "Sehr gegenseitiges Voting"
        elif asymmetrie < 2:
            interpretation = "Leicht asymmetrisch"
        elif asymmetrie < 4:
            interpretation = "Deutlich einseitig"
        else:
            interpretation = "Extrem einseitiges Voting"
        print(f"Interpretation: {interpretation}")
        print(f"Dominante Richtung: {dominante_richtung}")

    def zeige_top_votings(self, x=10):
        print(f"\nTOP {x} NACHBARSCHAFTSVOTINGS:\n")
        top_df = self.ergebnis.head(x)
        print(
            top_df[
                [
                    'Voter',
                    'Receiver',
                    'Jahre_Aktiv',
                    'Nachbarschafts_Bias'
                ]
            ].to_string(index=True)
        )

    def plot_matrix(self, laender_liste=None, top_x=12):
        from matplotlib.colors import LogNorm
        if laender_liste:
            iso_liste = [
                self._normalisiere_land(l)
                for l in laender_liste
            ]
        else:
            iso_liste = self.ergebnis['Voter'].unique()[:top_x]
        df_sub = self.ergebnis[ self.ergebnis['Voter'].isin(iso_liste) & self.ergebnis['Receiver'].isin(iso_liste) ]
        if df_sub.empty:
            print("Keine Daten vorhanden.")
            return
        pivot_df = (
            df_sub.pivot( 
                index='Voter',
                columns='Receiver',
                values='Nachbarschafts_Bias'
            ).fillna(1.0)
        )
        matrix = pivot_df.values.copy()
        np.fill_diagonal(matrix, np.nan)
        fig, ax = plt.subplots(figsize=(11, 9))
        cmap = plt.cm.coolwarm.copy()
        cmap.set_bad(color='black')
        cax = ax.imshow(
            matrix,
            cmap=cmap,
            norm=LogNorm(vmin=0.05, vmax=20.0)
        )
        cbar = fig.colorbar(
            cax,
            label='Nachbarschafts-Bias',
            ticks=[0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10, 20]
        )
        cbar.ax.set_yticklabels([
            '0.05x',
            '0.1x',
            '0.2x',
            '0.5x',
            '1x',
            '2x',
            '5x',
            '10x',
            '20x'
        ])
        ax.set_xticks(np.arange(len(pivot_df.columns)))
        ax.set_yticks(np.arange(len(pivot_df.index)))
        ax.set_xticklabels(
            pivot_df.columns,
            rotation=45,
            ha="right"
        )
        ax.set_xlabel(
            "Receiver (bekommt Punkte)",
            fontsize=11,
            fontweight='bold'
        )
        ax.set_ylabel(
            "Voter (vergibt Punkte)",
            fontsize=11,
            fontweight='bold'
        )
        ax.set_yticklabels(pivot_df.index)
        display_matrix = np.floor(matrix * 100) / 100
        for i in range(len(pivot_df.index)):
            for j in range(len(pivot_df.columns)):
                val = matrix[i, j]
                if np.isnan(val):
                    continue
                display_val = display_matrix[i, j]
                text_farbe = (
                    "white"
                    if val > 2.0
                    else "black"
                )
                ax.text(
                    j,
                    i,
                    f"{display_val:.2f}x",
                    ha="center",
                    va="center",
                    color=text_farbe,
                    fontsize=8,
                    fontweight='bold'
                )
        plt.title(
            "ESC Nachbarschafts-Bias Matrix",
            fontsize=14,
            fontweight='bold'
        )
        plt.tight_layout()
        plt.show()

esc = ESCAnalyzer(2016, 2025)
esc.zeige_paar_bias("WLD", "IL")
esc.zeige_paar_bias("AT", "WLD")
esc.zeige_paar_bias("GRE", "AT")
esc.zeige_einseitigkeit("WLD", "GB")
esc.zeige_einseitigkeit("AT", "DE")
esc.zeige_top_votings(x=5)
esc.plot_matrix(top_x=7)
esc.plot_matrix( laender_liste=[ "Deutschland", "Österreich", "Schweiz", "WLD" ] )
esc.plot_matrix( laender_liste=[ "Finnland", "Dänemark", "Norwegen", "Schweden", "Island" ] )
```
